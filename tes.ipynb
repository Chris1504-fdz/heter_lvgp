```python
import matplotlib.pyplot as plt

# Setup the figure dimensions
fig = plt.figure(figsize=(8, 4))

# The LaTeX equations formatted for Matplotlib's text engine
text = (
    r"$g(x) = \prod_{i=1}^{D} \sqrt{x_i} \sin(x_i)$" + "\n\n" +
    r"$h_1(x) = \sqrt{x_1} \sin(x_1)$" + "\n\n" +
    r"$h_k(x) = h_{k-1}(x) \cdot \sqrt{x_k} \sin(x_k) \quad \mathrm{for} \; k = 2, \dots, 6$" + "\n\n" +
    r"$g(x) = h_6(x)$"
)

# Add the text to the canvas
plt.text(0.05, 0.5, text, fontsize=18, va='center', ha='left')

# Hide the background axes
plt.axis('off')

# Save the output as a transparent PNG
plt.savefig("equations_transparent.png", transparent=True, bbox_inches='tight', dpi=300)
print("Success! Image saved as 'equations_transparent.png'")
```
